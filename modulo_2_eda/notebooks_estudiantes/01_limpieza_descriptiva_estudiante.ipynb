# 2. Limpieza de Presión y Caudal

## ¿Qué pozos reportan BHP = −999 psi, indicando una falla de calibración del transductor?

```python
# TU CÓDIGO AQUÍ: Encuentra las filas donde BHP es igual a -999.0 y agrúpalas por 'Pozo'
df[df['BHP_psi'] == -999.0].groupby('Pozo').size()
```

## ¿Cómo se ve la serie de tiempo de BHP del Pozo-01 antes de corregir la falla de telemetría?

```python
df_pozo1 = df[df['Pozo'] == 'Pozo-01']
plt.plot(df_pozo1['Fecha'], df_pozo1['BHP_psi'], label='BHP (Cruda)', color='red')
plt.title('BHP - Pozo-01 con anomalías de calibración')
plt.xlabel('Fecha')
plt.ylabel('BHP (psi)')
plt.legend()
plt.show()
```

## ¿Cómo reemplazamos los valores imposibles de −999 por valores nulos (NaN)?

```python
# TU CÓDIGO AQUÍ: Reemplaza los valores -999.0 de 'BHP_psi' por np.nan

print("Valores NaN en BHP:", df['BHP_psi'].isnull().sum())
```

## ¿Cómo imputamos esos valores usando interpolación lineal agrupada por cada pozo?

```python
# TU CÓDIGO AQUÍ: Interpola linealmente agrupando por la columna 'Pozo'
df['BHP_psi'] = None

print("Valores NaN residuales:", df['BHP_psi'].isnull().sum())
```

## ¿Cómo queda la serie temporal de BHP en el Pozo-01 tras aplicar la corrección?

```python
df_pozo1_limpio = df[df['Pozo'] == 'Pozo-01']
plt.plot(df_pozo1_limpio['Fecha'], df_pozo1_limpio['BHP_psi'], label='BHP (Corregida)', color='green')
plt.title('BHP - Pozo-01 después de interpolación lineal')
plt.xlabel('Fecha')
plt.ylabel('BHP (psi)')
plt.legend()
plt.show()
```
